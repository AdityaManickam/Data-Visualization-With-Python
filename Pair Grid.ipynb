```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
a = sns.load_dataset('tips')
```

```python
x = sns.PairGrid(a)
x = x.map(plt.scatter)
```

```python
x = sns.PairGrid(a)
x = x.map_diag(plt.hist)
```

```python
x = sns.PairGrid(a)
x = x.map_diag(plt.hist)
x = x.map_offdiag(plt.scatter)
```

```python
x = sns.PairGrid(a,hue='day')
x = x.map_diag(plt.hist)
x = x.map_offdiag(plt.scatter)
x = x.add_legend()
```

```python
x = sns.PairGrid(a,hue='sex')
x = x.map_diag(plt.hist,histtype='step')
x = x.map_offdiag(plt.scatter)
x = x.add_legend()
```

```python
x = sns.PairGrid(a,vars=['total_bill','tip'],hue='sex')
x = x.map_diag(plt.hist,edgecolor='black')
x = x.map_offdiag(plt.scatter)
```

```python
x = sns.PairGrid(a,x_vars=['total_bill','tip'],\
                 y_vars=['day','sex'])
x = x.map(plt.scatter,edgecolor='black')
```

```python
x = sns.PairGrid(a,hue='sex')
x = x.map_diag(plt.hist)
x = x.map_upper(plt.scatter)
x = x.map_lower(sns.kdeplot)
x = x.add_legend()
```

```python
x = sns.PairGrid(a,hue='sex',hue_kws={'marker':['D','s']})
x = x.map_diag(plt.hist)
x = x.map_offdiag(plt.scatter)
x = x.add_legend()
```
