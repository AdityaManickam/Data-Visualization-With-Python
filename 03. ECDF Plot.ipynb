```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
tips.shape
```

```python
sns.set_style('darkgrid')
```

```python
sns.ecdfplot(x='tip',data=tips)
```

```python
tips.tip.min()
```

```python
tips.tip.max()
```

```python
sns.ecdfplot(x='tip',data=tips)
plt.axvline(4,c='black')
```

```python
sns.ecdfplot(x='tip',data=tips,hue='time')
```

```python
sns.ecdfplot(x='tip',data=tips,hue='day')
```

```python
sns.ecdfplot(x='tip',data=tips,hue='time',stat='count') #Counting
```

```python
sns.ecdfplot(x='tip',data=tips,hue='time',stat='count',complementary='True') #Ranking
```

```python
sns.ecdfplot(x='tip',data=tips,weights='tip')
plt.axhline(0.5,c='black')
```

```python
p = sns.ecdfplot(x='tip', data=tips, hue='day', palette='summer', lw=3)
p.legend(['Thursday', 'Friday', 'Saturday', 'Sunday'], title='Day of Week', loc=4);
```
