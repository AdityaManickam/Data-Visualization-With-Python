```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
cars = sns.load_dataset('mpg').dropna()

cars.shape
```

```python
cars.head()
```

```python
sns.set_style('whitegrid')
```

```python
cars.cylinders.value_counts()
```

```python
cars = cars[cars.cylinders.isin([4,6,8])]
```

```python
sns.swarmplot(x='horsepower',data=cars);
```

```python
sns.swarmplot(x='origin', y='horsepower',data=cars);
```

```python
sns.swarmplot(x='origin', y='horsepower',hue='cylinders',data=cars);
```

```python
sns.swarmplot(x='origin', y='horsepower', hue='cylinders',data=cars, dodge=True);
```

```python
usa = cars[cars.origin=='usa']
```

```python
sns.boxplot(x='cylinders', y='horsepower',data=usa)
sns.swarmplot(x='cylinders', y='horsepower',data=usa);
```

```python
sns.boxplot(x=usa.cylinders, y=usa.horsepower,
            whis=np.inf)
sns.swarmplot(x=usa.cylinders, y=usa.horsepower,
              color='black', alpha=0.5);
```

```python
sns.violinplot(x=usa.cylinders, y=usa.horsepower,
               scale='width')
sns.swarmplot(x=usa.cylinders, y=usa.horsepower,
              color='white', size=4);
```

```python
sns.violinplot(x=usa.cylinders, y=usa.horsepower,
               scale='width', inner=None, color='xkcd:forest green' )
sns.swarmplot(x=usa.cylinders, y=usa.horsepower,
              color='white', size=3);
```

```python
sns.swarmplot(x='horsepower',data=cars, linewidth=1);
```

```python
sns.swarmplot(x='horsepower',data=cars, linewidth=1, edgecolor='black',marker='*');
```
