```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
cars = sns.load_dataset('mpg').dropna()

cars.shape
```

```python
cars.head()
```

```python
cars.cylinders.value_counts()
```

```python
cars = cars[cars.cylinders.isin([4,6,8])]
```

```python
sns.violinplot(x='displacement',data=cars);
```

```python
sns.violinplot(x='cylinders', y='displacement',data=cars);
```

```python
sns.violinplot(x='cylinders', y='displacement', hue='origin',data=cars);
```

```python
sns.violinplot(x='cylinders', y='displacement', hue='origin',
               data=cars[cars.origin.isin(['japan', 'europe'])],
               split=True)
plt.legend(loc=2);
```

```python
sns.violinplot(x='cylinders', y='displacement', hue='origin',
               data=cars[cars.origin.isin(['japan', 'europe'])],
               split=True,
               inner='quartiles')
plt.legend(loc=2);
```

```python
sns.violinplot(x='cylinders', y='displacement', hue='origin',
               data=cars[cars.origin.isin(['japan', 'europe'])],
               split=True,
               scale='count')
plt.legend(loc=2);
```

```python

sns.violinplot(x='cylinders', y='displacement', hue='origin',
               data=cars[cars.origin.isin(['japan', 'europe'])],
               split=True,
               scale='count',
               scale_hue=False)
plt.legend(loc=2);
```

```python

sns.violinplot(x=cars.origin, y=cars.displacement,
               order=['japan', 'europe', 'usa'],linewidth=3);
```

```python
sns.violinplot(x=cars.cylinders, y=cars.displacement,bw=0.2); #Bandwaidth
```
