```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
a = sns.load_dataset('tips')
```

```python
sns.stripplot(x=a['tip'], color='green')
```

```python
sns.stripplot(x=a['total_bill'], color='green',jitter=False)
```

```python
sns.stripplot(x='day',y='total_bill',data=a)
```

```python
sns.stripplot(x='day',y='total_bill',data=a, jitter=False)
```

```python
sns.stripplot(x='day',y='total_bill',data=a, jitter=0.5)
```

```python
sns.stripplot(x='total_bill',y='day',data=a,jitter=False )
```

```python
sns.stripplot(x='total_bill',y='day',data=a,jitter=1 )
```

```python
sns.stripplot(x='total_bill',y='day',data=a,jitter=True, linewidth=1.5 )
```

```python
sns.stripplot(x='day',y='total_bill',data=a,hue='sex', jitter=False)
```

```python
sns.stripplot(x='day',y='total_bill',data=a,hue='sex', jitter=True)
```

```python
sns.stripplot(x='day',y='total_bill',data=a,hue='sex', jitter=True, dodge=True)
```

```python
sns.stripplot(x='day',y='total_bill',data=a,hue='sex', jitter=False, marker='D')
```

```python
sns.stripplot(x='day',y='total_bill',data=a,hue='sex', jitter=False, marker='D',size=15,alpha=0.2)
```
