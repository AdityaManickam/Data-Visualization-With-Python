```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
a = sns.load_dataset('iris').dropna()
```

```python
a.shape
```

```python
a.head()
```

```python
sns.kdeplot(x='sepal_length',data=a)
```

```python
sns.kdeplot(x='sepal_length',data=a,shade=True)
```

```python
sns.kdeplot(x='sepal_length',data=a,cumulative=True) #Cumulative Density Function
```

```python
sns.kdeplot(x='sepal_length',y='sepal_width',data=a) #Bivariate
```

```python
sns.kdeplot(x='sepal_length',y='sepal_width',data=a, n_levels=20) 
```

```python
sns.kdeplot(x='sepal_length',y='sepal_width',data=a, n_levels=20,shade=True) 
```

```python
sns.set_style('whitegrid')
sns.kdeplot(x='sepal_length',y='sepal_width',data=a, n_levels=20,shade=True,thresh=False) 
```

```python
sns.set_style('whitegrid')
plt.figure(figsize=(8,6))

sns.kdeplot(x='sepal_length',y='sepal_width',data=a, n_levels=20,shade=True,thresh=False,cbar=True) 
```
