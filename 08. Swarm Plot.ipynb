```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
diamonds = sns.load_dataset('diamonds')
diamonds.shape
```

```python
diamonds.head()
```

```python
diamonds = diamonds[
            diamonds.cut.isin(['Premium', 'Good']) & 
            diamonds.color.isin(['D', 'F', 'J'])
            ].sample(n=100, random_state=22)
diamonds.shape
```

```python
sns.set_style('dark')
```

```python
sns.scatterplot(x='carat', y='price', data=diamonds);
```

```python
sns.scatterplot(x='x', y='y', data=diamonds);
```

```python
sns.set_style('darkgrid')
```

```python
sns.scatterplot(x='carat', y='price', hue='cut',
                data=diamonds,palette='husl')
```

```python
sns.scatterplot(x='carat', y='price', hue='cut',
                data=diamonds,
                palette = ['purple', '#55CCCC'],
                hue_order=['Good', 'Premium']);
```

```python
sns.scatterplot(x='carat', y='price', style='cut',
                data=diamonds,
                s=100
               );
```

```python
sns.scatterplot(x='carat', y='price', size='cut',
                data=diamonds);
```

```python
sns.scatterplot(x='carat', y='price', size='depth',
                data=diamonds);
```

```python
sns.scatterplot(x='carat', y='price', 
                hue='cut', 
                style='cut',
                data=diamonds,
                s=100
               );
```

```python
sns.scatterplot(x='carat', y='price', 
                hue='cut', 
                style='color',
                data=diamonds,
                s=100
               );
```

```python
sns.scatterplot(x='carat', y='price', 
                hue='cut', 
                style='color',
                size='depth',
                data=diamonds);
```

```python
sns.scatterplot(x='carat', y='price', 
                data=diamonds,
                s=100,
                alpha=0.6
               );
```

```python
sns.scatterplot(x='carat', y='price', 
                data=diamonds, 
                s=200,
                marker='*'
               );
```

```python
sns.scatterplot(x='carat', y='price', 
                data=diamonds, 
                s=200, 
                marker='*',
                edgecolor='black'
               );
```
