```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
cars = sns.load_dataset('mpg')
```

```python
cars.head()
```

```python
cars.groupby('origin').cylinders.value_counts()
```

```python
(cars
 .groupby('origin')
 .cylinders
 .value_counts()
 .unstack()
 .fillna(0)
)
```

```python
origin_cyl = (
    cars
    .groupby('origin')
    .cylinders
    .value_counts()
    .unstack()
    .fillna(0)
)
```

```python
sns.heatmap(origin_cyl);
```

```python
origin_cyl.T #Transpose
```

```python
sns.heatmap(origin_cyl.T); #Transpose
```

```python
cars.corr() #Correlation
```

```python
sns.heatmap(cars.corr()); #Correlation
```

```python
sns.heatmap(cars.corr(), cmap='RdBu');
```

```python
sns.heatmap(origin_cyl, cmap="Blues");
```

```python
sns.heatmap(origin_cyl, cmap='Blues', annot=True);
```

```python
sns.heatmap(origin_cyl, cmap='Blues', annot=True, fmt=".0f");  #fmt=".1f" for one decimal place
```

```python
sns.heatmap(origin_cyl, 
            cmap='Blues', 
            annot=True, 
            fmt=".0f",
            annot_kws={
                'fontsize': 16,
                'fontweight': 'bold',
                'fontfamily': 'serif',
               # 'color': 'black'
            }
           );
```

```python
sns.heatmap(cars.corr(), 
            cmap='RdBu', 
            linewidth=1,
            linecolor='black'
           );
```

```python
labels = ['MPG', 'Cylinders', 'Displacement', 'Horsepower',
          'Weight', 'Acceleration', 'Model Year'
         ]
```

```python
sns.heatmap(cars.corr(), 
            cmap='RdBu',
            xticklabels=labels,
            yticklabels=labels
           );
```

```python
sns.heatmap(cars.corr(), 
            cmap='RdBu',
            square=True
           ); #Square
```
