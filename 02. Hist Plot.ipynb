```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
penguins = sns.load_dataset('penguins')
```

```python
penguins.head()
```

```python
penguins.shape
```

```python
penguins.dropna(inplace=True)
```

```python
penguins.shape
```

```python
sns.set_style('darkgrid')
```

```python
sns.histplot(penguins.bill_length_mm)
```

```python
sns.histplot(x='bill_length_mm',data=penguins,kde=True)
```

```python
sns.histplot(x='bill_length_mm', data=penguins, bins=20); #Bins
```

```python
sns.histplot(x='bill_length_mm', data=penguins,bins=[30, 40, 45, 50, 60]);
```

```python
sns.histplot(x='bill_length_mm', data=penguins, 
             binwidth=5, binrange=(30, 60));
```

```python
sns.histplot(x='bill_length_mm', data=penguins,
             stat='count');
```

```python
sns.histplot(x='bill_length_mm', data=penguins,
             stat='density');
```

```python
sns.histplot(x='bill_length_mm', data=penguins,
             stat='probability');
```

```python
sns.histplot(x='bill_length_mm', data=penguins,stat='probability', fill=False, element='step',cumulative=True); #Cumulative
```

```python
sns.histplot(x='bill_length_mm', data=penguins, hue='species');
```

```python
sns.histplot(x='bill_length_mm', data=penguins, hue='species', element='step');
```

```python
sns.histplot(x='bill_length_mm', data=penguins, hue='species', element='poly');
```

```python
sns.histplot(x='bill_length_mm', data=penguins, hue='species',  multiple='stack');
```

```python
sns.histplot(x='bill_length_mm', data=penguins, hue='species',  multiple='fill');
```

```python
sns.histplot(x='species', data=penguins, hue='island',  multiple='dodge');
```

```python
sns.set_style('white')
```

```python
sns.histplot(x='bill_length_mm', y='bill_depth_mm', data=penguins); #Bivariate
```

```python
sns.histplot(x='bill_length_mm', y='bill_depth_mm', data=penguins,cbar=True);
```

```python
sns.histplot(x='bill_length_mm', y='bill_depth_mm',hue='species', data=penguins,cbar=True);
```

```python
sns.histplot(x='species',  hue='sex', multiple='dodge', data=penguins, shrink=0.8);
```

```python
sns.histplot(x='species',  hue='sex', multiple='dodge', data=penguins, shrink=0.8, palette='bone');
```

```python
sns.histplot(x='species',  hue='sex', multiple='dodge', data=penguins, shrink=0.8, palette='bone', fill=False);
```
