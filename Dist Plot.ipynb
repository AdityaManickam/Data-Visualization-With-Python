```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
num = np.random.randn(150)
```

```python
sns.distplot(num)
```

```python
sns.distplot(num, color='red')
```

```python
lable=pd.Series(num,name='Variable X')
```

```python
sns.distplot(lable)
```

```python
sns.distplot(lable, vertical=True)
```

```python
sns.distplot(lable, hist = False)
```

```python
sns.distplot(lable, hist = False,rug = True)
```
